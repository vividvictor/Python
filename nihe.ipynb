```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
#copyRight by heibanke 

import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline  
x = np.arange(-1,1,0.02)  
y = ((x*x-1)**3+1)*(np.cos(x*2)+0.6*np.sin(x*1.3)) #理想曲线

y1 = y+(np.random.rand(len(x))-0.5) #现有数据


### write your code to gen A,b
m=[]
for i in range(6):
    m.append(x**(i))
    
m.append(np.cos(x**2))
m.append(np.sin(x))

A=np.array(m).T
b=y1.reshape(y1.shape[0],1)

def projection(A,b):
    ####
    # return A*inv(AT*A)*AT*b
    ####
    AA = A.T.dot(A)
    w=np.linalg.inv(AA).dot(A.T).dot(b)
    print (w)
    return A.dot(w)

yw = projection(A,b)
yw.shape = (yw.shape[0],)

plt.plot(x,y,color='g',linestyle='-',marker='',label='best') 
plt.plot(x,y1,color='m',linestyle='',marker='o',label='data')
plt.plot(x,yw,color='r',linestyle='',marker='.',label='nihe')
plt.legend()
plt.show()
```
